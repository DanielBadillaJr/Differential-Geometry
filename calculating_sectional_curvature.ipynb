```python
x, mu, s = smp.symbols('x mu s', real=True)
pdf = 1/(s*smp.sqrt(2*np.pi))*smp.exp(-1/2*((x-mu)/s)**2)
pdf_times_x = pdf * x
pdf_times_x
```
